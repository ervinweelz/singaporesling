```python
import requests
from bs4 import BeautifulSoup

url_link = "https://www.channelnewsasia.com/singapore/iswaran-cpib-assist-investigations-ongoing-pm-lee-3622586"
result = requests.get(url_link).text

soup = BeautifulSoup(result, "html.parser")

# Find the parent element that contains the desired elements
parent_element = soup.find("article", class_="node node--article-content")

if parent_element is not None:
    print("going into parent element")
    # Use the parent element to find specific elements
    desired_elements = parent_element.find_all("div", class_="text-long")
    print("Number of desired elements found:", len(desired_elements))

    # Process the desired elements
#     for element in desired_elements:
#         # Extract and process the text or perform any other operations
#         function_name = element.get_text(strip=True)
        
#         print("my functon", function_name)
    paragraphs = parent_element.find_all("p")
    cleaned_text = " ".join([paragraph.get_text(strip=True) for paragraph in paragraphs])
    print(cleaned_text)
    
else:
    print("Parent element not found.")
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

# Assuming you have the 'cleaned_text' variable that contains the cleaned text

# Create an instance of CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the cleaned text into a document-term matrix
X = vectorizer.fit_transform([cleaned_text])

# Create an instance of TfidfTransformer
tfidf_transformer = TfidfTransformer()

# Transform the document-term matrix into a TF-IDF matrix
X_tfidf = tfidf_transformer.fit_transform(X)

# Get the feature names (words)
feature_names = vectorizer.get_feature_names()

# Convert X_tfidf to an array
X_tfidf_array = X_tfidf.toarray()

# Get the TF-IDF scores for the cleaned text
tfidf_scores = X_tfidf_array[0, :]

# Sort the indices of the scores in descending order
top_indices = tfidf_scores.argsort()[::-1]

# Extract the top 5 key words for the cleaned text
top_keywords = [feature_names[idx] for idx in top_indices[:5]]

# Print the key words for the cleaned text
print(f"Key words: {top_keywords}")
```
